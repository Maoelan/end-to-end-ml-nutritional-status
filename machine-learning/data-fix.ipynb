```python
import pandas as pd
import numpy as np
```

```python
data_c = pd.read_excel('data clean.xlsx')
data_c
```

```python
data_bulan = data_c['Usia Saat Ukur']
data_bulan
```

```python
def hitungan_bulan(data):
    hasil = []
    for d in data:
        # Pisahkan bagian tahun, bulan, dan hari
        bagian = d.split(" - ")
        tahun = int(bagian[0].split(" TAHUN")[0])
        bulan = int(bagian[1].split(" BULAN")[0])
        hari = int(bagian[2].split(" HARI")[0])

        # Konversi ke hitungan bulan
        total_bulan = tahun * 12 + bulan

        hasil.append(total_bulan)
    
    return hasil
```

```python
data_c['Usia Saat Ukur'] = hitungan_bulan(data_c['Usia Saat Ukur'])
data_c
```

```python
data_c.to_excel('data final.xlsx', index=False)
```
