```python
def convert_to_days(age):
    parts = age.split('-')
    years = 0
    months = 0
    days = 0
    for part in parts:
        part = part.strip()
        if 'Tahun' in part:
            years = int(part.split(' ')[0])
        elif 'Bulan' in part:
            months = int(part.split(' ')[0])
        elif 'Hari' in part:
            days = int(part.split(' ')[0])
    total_days = years * 365 + months * 30 + days
    return total_days
```

```python
new_data['Usia Saat Ukur'] = new_data['Usia Saat Ukur'].apply(convert_to_days)
```
