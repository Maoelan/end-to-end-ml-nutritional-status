```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('data training.xlsx')
data
```

```python
data.describe()
```

```python
data['RT'] = data['RT'].fillna(0)
data['RW'] = data['RW'].fillna(0)
```

```python
data.isnull().sum()
```

```python
data['Jml Vit A'] = data['Jml Vit A'].fillna(0)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
def hurufbesar(df):
    data = df.select_dtypes(include=['object'])
    df[data.columns] = data.apply(lambda x: x.str.upper())
    return df
```

```python
data = hurufbesar(data)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
data.info()
```

```python
data['BB Lahir'] = data['BB Lahir'].astype('float64')
data['TB Lahir'] = data['TB Lahir'].astype('float64')
```

```python
data
```

```python
data['RT'] = data['RT'].astype('int64')
data['RW'] = data['RW'].astype('int64')
data['Jml Vit A'] = data['Jml Vit A'].astype('int64')
```

```python
data['RT'] = data['RT'].astype('str')
data['RW'] = data['RW'].astype('str')
```

```python
data
```

```python
data.describe()
```

```python
data.info()
```

```python
numerik_feature = ['BB Lahir', 'TB Lahir', 'Berat', 'Tinggi', 'Jml Vit A']
```

```python
sns.boxplot(data=data[numerik_feature])
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

IQR = Q3 - Q1

data = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]
```
